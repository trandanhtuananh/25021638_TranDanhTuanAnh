```python
#W11A1
def reverse(path: str) -> int:
    try:
        with open(path, 'r') as f:
            n_str = f.read().strip()
            return int(n_str[::-1])
    except:
        return 0

path = input()
print(reverse(path))
```

```python
#W11A2
def canWinNim(path: str) -> bool:
    try:
        with open(path, 'r') as f:
            n = int(f.read().strip())
            return n % 4 != 0
    except:
        return False

path = input()
print(canWinNim(path))
```

```python
#W11A3
def housesOfHogwarts(path: str):
    try:
        with open(path, 'r') as f:
            lines = f.readlines()
            if lines:
                n = int(lines[0].strip())
                for i in range(1, n + 1):
                    print(lines[i].strip())
    except:
        pass

path = input()
housesOfHogwarts(path)
```

```python
#W11A4
import re
def areAnagrams(path1: str, path2: str) -> bool:
    def clean(text):
        return sorted(re.sub(r'[^a-z]', '', text.lower()))
    
    try:
        with open(path1, 'r') as f1, open(path2, 'r') as f2:
            return clean(f1.read()) == clean(f2.read())
    except:
        return False

p1 = input()
p2 = input()
print(areAnagrams(p1, p2))
```

```python
#W11A5 & W11A6
def countCharacter(path: str, letter: str) -> int:
    try:
        with open(path, 'r') as f:
            content = f.read().lower()
            return content.count(letter.lower())
    except:
        return 0

path = input()
char = input()
print(countCharacter(path, char))
```

```python
#W11A10
def getMoney(path: str) -> list:
    with open(path, 'r') as f:
        return list(map(int, f.read().split()))

def rob(houses: list) -> int:
    prev2, prev1 = 0, 0
    for x in houses:
        prev2, prev1 = prev1, max(prev1, prev2 + x)
    return prev1

path = input()
money = getMoney(path)
print(rob(money))
```

```python
#W11A12
def maximumProduct(path: str) -> int:
    with open(path, 'r') as f:
        lines = f.readlines()
        nums = sorted(map(int, lines[1].split()))
        return max(nums[-1]*nums[-2]*nums[-3], nums[0]*nums[1]*nums[-1])

path = input()
print(maximumProduct(path))
```

```python
#W11A13
def moveZeroes(fileName: str) -> list:
    with open(fileName, 'r') as f:
        nums = list(map(int, f.readlines()[1].split()))
        res = [x for x in nums if x != 0]
        res.extend([0] * (len(nums) - len(res)))
        return res

path = input()
print(moveZeroes(path))
```

```python
#W11A14
import csv
def averageTime(path: str):
    try:
        with open(path, 'r') as f:
            reader = csv.DictReader(f)
            times = [float(row['Time']) for row in reader]
            avg = sum(times) / len(times)
            print(f"{avg:.2f}")
    except:
        pass

path = input()
averageTime(path)
```

```python
#W11A15
import csv
def findMovies(path: str):
    res = []
    try:
        with open(path, 'r') as f:
            reader = csv.DictReader(f)
            for row in reader:
                if float(row['rating']) > 8.0 and "boring" not in row['description'].lower():
                    res.append(row['movie'].strip())
        if res:
            for m in res: print(m)
        else:
            print("All are boring!")
    except:
        print("All are boring!")

path = input()
findMovies(path)
```

```python
#W11A16
import math
path = input()
try:
    with open(path, 'r') as f:
        content = f.read().split('TYPE: ')
        for item in content[1:]:
            lines = item.strip().split('\n')
            shape = lines[0].strip()
            params = {p.split(': ')[0]: float(p.split(': ')[1]) for p in lines[1].split(' ')}
            
            if shape == 'SQUARE':
                print(f"{4 * params['SIDE']:.2f}")
            elif shape == 'RECTANGLE':
                print(f"{2 * (params['WIDTH'] + params['HEIGHT']):.2f}")
            elif shape == 'CIRCLE':
                print(f"{2 * math.pi * params['RADIUS']:.2f}")
except:
    pass
```

```python
#W11A17
def productExceptSelf(path: str) -> list:
    with open(path, 'r') as f:
        nums = list(map(int, f.read().split()))
    n = len(nums)
    res = [1] * n
    left = 1
    for i in range(n):
        res[i] = left
        left *= nums[i]
    right = 1
    for i in range(n-1, -1, -1):
        res[i] *= right
        right *= nums[i]
    return res

path = input()
print(productExceptSelf(path))
```

```python
#W11A18
import math
def estimatedTime(path: str):
    try:
        with open(path, 'r') as f:
            lines = f.readlines()
            src = [float(v) for v in re.findall(r'-?\d+\.?\d*', lines[0])]
            des = [float(v) for v in re.findall(r'-?\d+\.?\d*', lines[1])]
            vel = float(re.findall(r'\d+\.?\d*', lines[2])[0])
            dist = math.sqrt((des[0]-src[0])**2 + (des[1]-src[1])**2)
            print(f"{dist/vel:.2f}")
    except:
        pass

path = input()
estimatedTime(path)
```

```python
#W11A19
import struct
path = input()
try:
    with open(path, 'rb') as f:
        while True:
            chunk = f.read(8)
            if not chunk: break
            val = struct.unpack('d', chunk)[0]
            print(f"{val:.5f}")
except:
    pass
```

```python
#W11A20
def imageSmoother(path: str):
    with open(path, 'r') as f:
        lines = f.readlines()
        M = int(lines[0].strip())
        img = [list(map(int, l.split())) for l in lines[1:M+1]]
    
    res = [[0]*M for _ in range(M)]
    for r in range(M):
        for c in range(M):
            vals = []
            for i in range(max(0, r-1), min(M, r+2)):
                for j in range(max(0, c-1), min(M, c+2)):
                    vals.append(img[i][j])
            res[r][c] = sum(vals) // len(vals)
    
    for row in res:
        print(*(row))

path = input()
imageSmoother(path)
```
