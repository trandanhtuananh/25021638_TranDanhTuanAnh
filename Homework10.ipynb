```python
#W9A1
l1 = list(map(int, input().split()))
l2 = list(map(int, input().split()))
i = j = 0
res = []
while i < len(l1) and j < len(l2):
    if l1[i] < l2[j]:
        res.append(l1[i])
        i += 1
    else:
        res.append(l2[j])
        j += 1
print(*(res + l1[i:] + l2[j:]))
```

```python
#W9A3
s = list(input().strip())
n = len(s)
res = "".join(s)
for i in range(n):
    for j in range(i + 1, n):
        s[i], s[j] = s[j], s[i]
        curr = "".join(s)
        if curr[0] != '0':
            if curr < res:
                res = curr
        s[i], s[j] = s[j], s[i]
print(res)
```

```python
#W9A4
q = int(input())
for _ in range(q):
    a, b = input(), input()
    i = j = 0
    possible = True
    while i < len(a) and j < len(b):
        if a[i].upper() == b[j]:
            i += 1
            j += 1
        elif a[i].isupper():
            possible = False
            break
        else:
            i += 1
    while i < len(a):
        if a[i].isupper():
            possible = False
        i += 1
    print("YES" if possible and j == len(b) else "NO")
```

```python
#W9A6
def next_permutation(arr):
    i = len(arr) - 2
    while i >= 0 and arr[i] >= arr[i+1]:
        i -= 1
    if i >= 0:
        j = len(arr) - 1
        while arr[j] <= arr[i]:
            j -= 1
        arr[i], arr[j] = arr[j], arr[i]
    arr[i+1:] = reversed(arr[i+1:])
    return arr

n = int(input())
arr = list(map(int, input().split()))
print(*(next_permutation(arr)))
```

```python
#W9A7
f = list(map(int, input()))
k = int(input())
c = 0
for i in range(len(f)):
    if f[i] == 0 and (i == 0 or f[i-1] == 0) and (i == len(f)-1 or f[i+1] == 0):
        f[i] = 1
        c += 1
print(c >= k)
```

```python
#W9A8
m = input()
x, y, seen, cross = 0, 0, {(0,0)}, False
for move in m:
    if move == 'U':
        y += 1
    elif move == 'D':
        y -= 1
    elif move == 'L':
        x -= 1
    elif move == 'R':
        x += 1
    if (x,y) in seen:
        cross = True
        break
    seen.add((x,y))
print(cross)
```

```python
#W9A10
import bisect
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()
    res = 0
    for h in houses:
        idx = bisect.bisect_left(heaters, h)
        dist1 = abs(heaters[idx] - h) if idx < len(heaters) else float('inf')
        dist2 = abs(heaters[idx-1] - h) if idx > 0 else float('inf')
        res = max(res, min(dist1, dist2))
    return res

h_list = list(map(int, input().split()))
ht_list = list(map(int, input().split()))
print(findRadius(h_list, ht_list))
```
