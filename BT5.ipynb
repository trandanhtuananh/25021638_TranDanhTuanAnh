```python
#W5A3
n = int(input("Nhap so n: "))
def snt(k):
    if k == 2 or k == 3:
        return True
    for i in range(2, int(k**(0.5)+1)):
        if k % i == 0:
            return False
    return True
if snt(n) == True:
    print("La so nguyen to")
else:
    print("Khong la so nguyen to")
```

```python
#W5A4
n = int(input("Nhap so n: "))
def snt(s):
    if s == 2 or s == 3:
        return True
    for i in range(2, int(s**(0.5)+1)):
        if s % i == 0:
            return False
    return True
def so_hoan_hao(k):
    if snt(k) == True:
        return False
    tong = 1
    for l in range(2, int(k**(0.5)+1)):
        if k % l == 0:
            tong += l + int(k//l)
    if tong == k:
        return True
    return False
if so_hoan_hao(n) == True:
    print("Day la so hoan hao")
else:
    print("Day khong phai la so hoan hao")
```

```python
#W5A5
k = int(input("Nhap so k: "))
lst = list(map(int, input("Nhap 1 day so nguyen: ").split()))
def vtri(n):
    global lst
    if n in lst:
        return lst.index(n)+1
    else:
        return -1
print(vtri(k))
        
```

```python
#W5A6
n = int(input("Nhap n: "))
def giai_thua(k):
    gt = 1
    for i in range(2, k + 1):
        gt *= i
    return gt
print(giai_thua(n))
```

```python
#W5A7
num1, num2 = map(int, input("Nhap 2 so num1 va num2: ").split())
pheptoan = int(input("1. Tru \n2. Cong \n3.Chia \n4.Nhan \nChon:"))
def mtbt(a, b):
    match pheptoan:
        case 1:
            return a - b
        case 2:
            return a + b
        case 3:
            return format(a/b, ".2f")
        case 4:
            return a * b
print(mtbt(num1, num2))
```

```python
#W5A8
m, n = map(int, input("Nhap 2 so m va n: ").split())
def kcach_Hamming(a, b):
    count = 0
    a = format(a, "b")
    b = format(b, "b")
    if len(a) > len(b):
        b = str(b).zfill(len(a))
    else:
        a = str(a).zfill(len(b))
    lsta = list(a)
    lstb = list(b)
    if a == b:
        return 0
    for k, i in zip(lsta, lstb):
        if k != i:
            count += 1
    return count
print(kcach_Hamming(m, n))
```

```python
#W5A9
n = int(input("Nhap so n: "))
def tcs(k):
    tong = 0
    while k > 0:
        tong += k % 10
        k //= 10
    return tong
print(tcs(n))
```
