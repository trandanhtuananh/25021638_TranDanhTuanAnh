```python
#W7A1
def binary_search(arr,value):
    left=0
    right=len(arr)-1
    while left <= right:
        mid=(left+right)//2
        if arr[mid]==value:
            return mid
        elif arr[mid]<value:
            left=mid+1
        else:
            right=mid-1
    return -1

N=list(map(int,input().split()))
x=int(input())
print(binary_search(N,x))
```

```python
#W7A2
def count_occurence(arr,x):
    count=0
    for i in arr:
        if i==x:
            count+=1
    return count

N=list(map(int,input().split()))
x=int(input())
print(count_occurence(N,x))
```

```python
#W7A7
import ast
line = input().strip()
nums = ast.literal_eval(line) if line.startswith('[') else list(map(int, line.split()))
dp = [1] * len(nums)
for i in range(len(nums)):
    for j in range(i):
        if nums[i] > nums[j]:
            dp[i] = max(dp[i], dp[j]+1)
print(max(dp) if nums else 0)
```

```python
#W7A8
import ast
intervals = ast.literal_eval(input())
queries = ast.literal_eval(input())
res = []
for q in queries:
    m_len = -1
    for l, r in intervals:
        if l <= q <= r:
            length = r - l + 1
            if m_len == -1 or length < m_len:
                m_len = length
    res.append(m_len)
print(res)
```

```python
#W7A9
import ast
line = input().strip()
nums = ast.literal_eval(line) if line.startswith('[') else list(map(int, line.split()))
res = []
for i in range(len(nums)):
    count = 0
    for j in range(i+1, len(nums)):
        if nums[j] < nums[i]:
            count += 1
    res.append(count)
print(res)
```

```python
#W7A10
import bisect
nums, m = eval(input()), int(input())
prefix, max_v, seen = 0, 0, [0]
for x in nums:
    prefix = (prefix + x) % m
    idx = bisect.bisect_right(seen, prefix)
    if idx < len(seen):
        max_v = max(max_v, (prefix - seen[idx] + m) % m)
    else:
        max_v = max(max_v, prefix)
    bisect.insort(seen, prefix)
print(max_v)
```
