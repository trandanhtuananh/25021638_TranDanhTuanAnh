```python
#W6A17
n = int(input("Nhap n: "))
mangip = []
dcc = ""
dcp = ""
flag = True
for i in range(1, n + 1):
    mang = list(map(int, input("Nhap hang: ").split()))
    if len(mang) == n:
        mangip.append(mang)
    else:
        flag = False
        break
if flag == True:
    for k in range(n):
        dcc += str(mangip[k][k]) + " "
        dcp += str(mangip[k][n-k-1]) + " "
    print(dcc)
    print(dcp)
else:
    print("Sai input!")
```

```python
#W6A18
n, m, k = map(int, input("Nhap n, m, k: ").split())
mangip = []
tongc = 0
tongh = 0
flag = True
for i in range(1, n + 1):
    mang = list(map(int, input("Nhap hang: ").split()))
    if len(mang) == m:
        mangip.append(mang)
    else:
        flag = False
        break
if flag == True:
    if k > n and k > m:
        print("Khong ton tai hang k va cot k")
    elif k > n:
        for s in range(n):
            tongc += mangip[s][k-1]
        print(str(tongc))
        print("Khong ton tai hang k")
    elif k > m:
        for s in range(m):
            tongh += mangip[k-1][s]
        print(str(tongh))
        print("Khong ton tai cot k")
    else:
        for s in range(n):
            tongc += mangip[s][k-1]
            tongh += mangip[k-1][s]
        print(str(tongc) + " " + str(tongh))
else:
    print("Sai input!")
```

```python
#W6A19
#19 va 22 trung nhau nen thay doi 19 thanh gia tri be nhat trong list
lst = list(map(int, input("Nhap list so nguyen: ").split()))
def bnhat(day):
    vtr = 0
    sbn = day[0]
    for i in range(len(day)):
        if day[i] <= sbn:
            sbn = day[i]
            vtr = i + 1
    return str(sbn) + " " + str(vtr)
print(bnhat(lst))
```

```python
#W6A20
lst = list(map(int, input("Nhap mot list so nguyen: ").split()))
k = int(input("Nhap so nguyen k: "))
def ptdt(day):
    for s in range(len(day)):
        if day[s] == k:
            return s + 1
    return -1
print(ptdt(lst))
```

```python
#W6A22
lst = list(map(int, input("Nhap 1 list so nguyen: ").split()))
def lnhat(day):
    sln = 0
    vtri = 0
    for i in day:
        if i >= sln:
            sln = i
            vtri = day.index(i) + 1
    return str(sln) + " " + str(vtri)
print(lnhat(lst))
```

```python
#W6A23
lst = list(map(int, input("Nhap mang gom cac so nguyen: ").split()))
x = int(input("Nhap x: "))
def xoax(day):
    dich = 0
    while dich <= len(day)-1:
        if day[dich] < x:
            day.remove(day[dich])
            dich = 0
        else:
            dich += 1
    return day
print(xoax(lst))
```

```python
#W6A24
lst = list(map(int, input("Nhap cac so nguyen: ").split()))
ans = ""
for k in range(len(lst)):
    if lst[k] == 0:
        lst.insert(0, lst.pop(k))
for i in lst:
    ans += str(i) + " "
print(ans)
```
