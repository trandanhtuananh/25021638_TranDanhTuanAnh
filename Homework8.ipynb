```python
#W7A16
from collections import Counter
arr = list(map(int, input().split()))
c = Counter(arr)
rem = 0
for x, count in c.items():
    if x == 0:
        rem += count
    elif count > x:
        rem += (count - x)
    elif count < x:
        rem += count
if rem == 0:
    print("YES")
else:
    print(f"NO {rem}")
```

```python
#W7A17
arr = list(map(int, input().split()))
cand, count = None, 0
for x in arr:
    if count == 0:
        cand, count = x, 1
    elif x == cand:
        count += 1
    else:
        count -= 1
if arr.count(cand) > len(arr)//2:
    print(cand)
```

```python
#W7A27
s = input()
stack, m = [], {')':'(', '}':'{', ']':'['}
valid = True
for c in s:
    if c in '({[':
        stack.append(c)
    elif c in m:
        if not stack or stack.pop() != m[c]:
            valid = False
            break
print("YES" if valid and not stack else "NO")
```

```python
#W7A28
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

n = int(input())
curr, longest = [], []
for i in range(2, n + 1):
    if not is_prime(i):
        curr.append(i)
    else:
        if len(curr) > len(longest):
            longest = curr
        curr = []
final = curr if len(curr) > len(longest) else longest
print(*(final))
```

```python
#W7A29
arr = list(map(int, input().split()))
curr, best = [arr[0]], [arr[0]]
for i in range(1, len(arr)):
    if arr[i] > arr[i-1]:
        curr.append(arr[i])
    else:
        if len(curr) > len(best):
            best = curr
        curr = [arr[i]]
final = curr if len(curr) > len(best) else best
print(*(final))
```

```python
#W7A30
def length_of_lis(nums):
    if not nums:
        return 0
    dp = [1] * len(nums)
    for i in range(len(nums)):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)

arr = list(map(int, input().split()))
print(length_of_lis(arr))
```

```python
#W7A31
s = list(input().strip())
n = len(s)
min_s = "".join(s)
for i in range(n):
    for j in range(i + 1, n):
        s[i], s[j] = s[j], s[i]
        curr = "".join(s)
        if curr[0] != '0' or len(curr) == 1:
            if curr < min_s:
                min_s = curr
        s[i], s[j] = s[j], s[i]
print(min_s)
```
