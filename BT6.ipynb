```python
#W6A9
ip = input("Nhap vao dictionary thu 1: ").strip()
ip2 = input("Nhap vao dictionary thu 2: ").strip()
def taodic(td):
    dic = {}
    tdms = list(map(str, td.strip('{}').split(',')))
    for p in tdms:
        key, value = p.split(':')
        value = value.strip()
        key = key.strip().strip("'")
        if key not in dic:
            dic[key] = []
        dic[key].append(value)
    return dic
if ip == '{}':
    ip2dic = taodic(ip2)
    print(dict(sorted(ip2dic.items())))
elif ip2 == '{}':
    ipdic = taodic(ip)
    print(dict(sorted(ipdic.items())))
else:
    ipdic = taodic(ip)
    ip2dic = taodic(ip2)
    for i in ip2dic:
        if i in ipdic:
            ipdic[i] += ip2dic[i]
    ipdic = dict(sorted(ipdic.items()))
    print(ipdic)
```

```python
#W6A10
lst = list(map(int, input("Nhap vao dsach cac so nguyen: ").split()))
k = int(input("Nhap so k: "))
tup = []
for i in range(len(lst)):
    tong = 0
    for l in range(i + 1, len(lst)):
        tong = lst[l] + lst[i]
        if tong == k and (lst[l], lst[i]) not in tup:
            tup.append((lst[i], lst[l]))
            
tup = sorted(tup)
print(tup)
```

```python
#W6A12
lst = list(map(int, input("Nhap mot list cac so nguyen: ").split()))
dic = {}

for i in lst:
    if i in dic:
        dic[i] += 1
    else:
        dic[i] = 1
so, sln = max(dic.items(), key= lambda x: x[1])
for k in dic:
    if dic[k] == sln:
        if k < so:
            so = k
print(so)
```

```python
#W6A15
ip = input("Nhap mot dictionary: ").strip()
k = int(input("Nhap k: "))
lst = list(map(str, ip.strip("{}").split(",")))
dic = {}
for p in lst:
    key, value = p.split(":")
    key = key.strip().strip("'")
    value = int(value)
    if value > k:
        if key not in dic:
            dic[key] = []
        dic[key].append(value)
print(dic)
        
```
